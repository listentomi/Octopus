```python
import json 


with open('cspider_examples.json', 'r') as f:
    cspider = json.load(f)

with open('spider_examples.json', 'r') as f:
    spider = json.load(f)

en2ch_examples = []
for crow,srow in zip(cspider,spider):
    print(crow['question'],srow['question'])
    en2ch_examples.append({
        'chinese question': crow['question'],
        'english question': srow['question'],
    })

file_path = 'en2ch_examples.json'
with open(file_path, 'a') as f:
    json.dump(en2ch_examples, f, ensure_ascii=False)  
```

```python
## 提取cspider问答对样本


import json 


with open('train.json', 'r') as f:
    data = json.load(f)

new_json = []
for item in data:
    question = item['question']
    sql = item['query']
    new_json.append({
        'question': question,
        'sql': sql
    })

file_path = 'cspider_examples.json'

with open(file_path, 'a') as f:
    json.dump(new_json, f, ensure_ascii=False)  
```
